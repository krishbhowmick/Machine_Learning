```
#seperating independent and dependent variables
x = mydata.iloc[:,[1]]
y = mydata['Math']
x.shape, y.shape
```

```

from sklearn.linear_model import LinearRegression
```

```

trainer=LinearRegression(fit_intercept=True,normalize=False)
```

```
learner=trainer.fit(x,y)
```

```
#learner.predict([[64,59]])
```

```
train_predict = learner.predict(x)
#train_predict
```

```
from sklearn.metrics import mean_absolute_error as mae
```

```
k = mae(train_predict, y)
print('Training Mean Absolute Error', k )
```

```
# Input
plt.plot(x,y,'r*')

# Predicted
#plt.plot(x,train_predict,'b*')
# Regression Line
plt.plot(x,train_predict,'g')
plt.xlabel("Input")
plt.ylabel("Ya")
#plt.legend(["Ya","Yp"])
plt.show()
```
